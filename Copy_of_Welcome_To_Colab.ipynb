```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/wpl2025_data.csv")

# Count the number of wins for each team
win_counts = df['Winning Team'].value_counts()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(win_counts, labels=win_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Match Wins by Team - WPL 2025')
plt.axis('equal')
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your dataset
df = pd.read_csv('/content/drive/MyDrive/wpl2025_data.csv')  # Replace with your file name

# Step 2: Select numerical columns only
numerical_cols = df.select_dtypes(include=['number']).columns

# Optional: Show sum of values for each column
print(df[numerical_cols].sum())

# Step 3: Create pie chart of total values from each numerical column
df[numerical_cols].sum().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    figsize=(6, 6),
    title="Numerical Columns Distribution"
)

plt.ylabel('')  # Hide y-axis label
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/wpl2025_data.csv')

# Scatter plot between two numerical columns
plt.scatter(df['first_innings_score'], df['second_innings_score'])
plt.title('first_innings_score vs second_innings_score')
plt.xlabel('first_innings_score')
plt.ylabel('second_innings_score')
plt.grid(True)
plt.show()
```
